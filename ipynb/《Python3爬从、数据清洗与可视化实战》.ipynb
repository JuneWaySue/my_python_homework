```python
url='https://cdn.heweather.com/china-city-list.txt'
headers={'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.100 Safari/537.36'}
r=requests.get(url)
data=r.content.decode()
data1=data.split('\n')
city_id=[]
for i in range(6):
    data1.remove(data1[0])
for each in data1:
    city_id.append(each[2:13])  #获取城市代码ID

city_list=[]
shijian_list=[]
weather_list=[]
wendu_list=[]
for i in range(1000):
    print('爬到第{0}个城市'.format(i+1))
    link='https://free-api.heweather.net/s6/weather/now?location='+city_id[i]+'&key=bed4a4c8c8af499588ed71e4d836568e'
    res=requests.get(link,headers=headers)
    j=res.json()
    city=j['HeWeather6'][0]['basic']['parent_city']
    shijian=j['HeWeather6'][0]['update']['loc']
    weather=j['HeWeather6'][0]['now']['cond_txt']
    wendu=j['HeWeather6'][0]['now']['tmp'] + '℃'
    city_list.append(city)
    shijian_list.append(shijian)
    weather_list.append(weather)
    wendu_list.append(wendu)
    time.sleep(1)
    
columns=['时间','城市','天气','温度']
df=pd.DataFrame(data=[shijian_list,city_list,weather_list,wendu_list]).T
df.columns=columns
df.to_excel('D:/python3.7/HEHE/输出/全国当前天气情况.xlsx',index=False)
```

```python
link='https://touch.dujia.qunar.com/depCities.qunar'
headers={'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.100 Safari/537.36'}
res=requests.get(link,headers=headers)
j=res.json()
start_city=[]  #获取出发点城市
end_city=[]    #获取目的地城市
for each in j['data']:
    for eachone in j['data'][each]:
        start_city.append(eachone)
        
for i,city in enumerate(start_city):
    print('爬到第{0}个城市'.format(i+1))
    hehe=urllib.request.quote(city)
    url='https://touch.dujia.qunar.com/golfz/sight/arriveRecommend?dep={0}&exclude=&extensionImg=255,175'.format(hehe)
    r=requests.get(url,headers=headers)
    jj=r.json()
#     time.sleep(random.randint(0,1)+random.random())
    for each in jj['data'][1:]:
        for sub in each['subModules']:
            for item in sub['items']:
                if item['query'] not in end_city:
                    end_city.append(item['query'])
```

## 模仿网页自动化爬取

```python
option=webdriver.ChromeOptions()
option.add_argument('disable-infobars') #把Chrome的警示关掉
d=webdriver.Chrome(options=option)
city_list=['广州']
results=[]
for city in city_list:
    hehe=urllib.request.quote(city)
    r=requests.get('https://touch.dujia.qunar.com/golfz/sight/arriveRecommend?dep={0}&exclude=&extensionImg=255,175'.format(hehe))
    j=r.json()
    time.sleep(2)
    for each in j['data'][1:]:
        for sub in each['subModules']:
            for item in sub['items']:
                query=item['query']
                d.get('https://fh.dujia.qunar.com/?tf=package')
                WebDriverWait(d,10).until(EC.presence_of_element_located((By.ID,"depCity")))
                d.find_element_by_xpath('//*[@id="depCity"]').clear()
                d.find_element_by_xpath('//*[@id="depCity"]').send_keys(city)
                d.find_element_by_xpath('//*[@id="arrCity"]').send_keys(query)
                d.find_element_by_xpath('/html/body/div[2]/div[1]/div[2]/div[3]/div/div[2]/div/a').click()
                print('出发点:{},目的地:{}'.format(city,query))
                for i in range(10):
                    time.sleep(random.uniform(8,10))
                    routes=d.find_elements_by_xpath('//*[@id="list"]/div')
                    if routes == []:
                        break
                    for route in routes:
                        result={'日期': time.strftime('%Y-%m-%d',time.localtime()),'出发点':city,'目的地':query,'内容':route.text}
                        results.append(result)
                    if i<9 :
                        btns=d.find_elements_by_xpath('//*[@id="pager"]/div/a')
                        for b in btns:
                            if b.text == '下一页':
                                b.click()
                                time.sleep(random.uniform(8,10))
                                break
```

```python
df=pd.DataFrame(data=results)
df.to_excel('D:/python3.7/HEHE/输出/在广州出发的旅游.xlsx',index=False)
```

```python
url_list=[]
title_list=[]
price_list=[]
day_list=[]
leixing_list=[]
d=webdriver.Chrome()
d1=webdriver.Chrome()
d.get('https://touch.dujia.qunar.com/p/list?it=n_index_desti_recom&dep=%E6%B1%9F%E9%97%A8&query=%E4%B8%BD%E6%B1%9F&das=1')
time.sleep(5)
for i in range(50):
    print('滚第{}下'.format(i+1))
    ActionChains(d).send_keys(Keys.PAGE_DOWN).perform()
    time.sleep(2)
a=d.find_elements_by_css_selector('div.item.g-flexbox.list-item ')
for i,each in enumerate(a[1:]):
    url=each.get_attribute('data-url')
    url_list.append(url)
    d1.get(url)
    print('爬到第{}个,url:{}'.format(i+1,url))
    time.sleep(5)
    for i in range(5):
        ActionChains(d1).send_keys(Keys.PAGE_DOWN).perform()
    try:
        if 'fhtouch' not in url:
            title=d1.find_element_by_css_selector('#root > div > div > div.flex.scrollable > div > div.main > div.m-hdinfo > div.title')
            title_list.append(title.text)
            price=d1.find_element_by_css_selector('#root > div > div > div.flex.scrollable > div > div.main > div.m-hdinfo > div.price > div > div.flex.g-flexbox > span > span.num')
            price_list.append(price.text)
            day=d1.find_element_by_css_selector('#root > div > div > div.flex.scrollable > div > div.main > div.m-summary > div:nth-child(2) > div.d.flex')
            day_list.append(day.text)
            leixing=d1.find_element_by_css_selector('#root > div > div > div.flex.scrollable > div > div.main > div.m-summary > div:nth-child(4) > div.d.flex')
            leixing_list.append(leixing.text)
            time.sleep(1)
        else:
            title=d1.find_element_by_css_selector('#header-info > div > div.m-hdinfo > div.title')
            title_list.append(title.text)
            price=d1.find_element_by_css_selector('#price-per')
            price_list.append(price.text)
            day=d1.find_element_by_css_selector('#js-baseInfo > div > div:nth-child(2) > div.d.flex')
            day_list.append(day.text)
            leixing=d1.find_element_by_css_selector('#js-baseInfo > div > div:nth-child(3) > div.d.flex.js-baseinfo-hotelDesc')
            leixing_list.append(leixing.text)
            time.sleep(1)
    except Exception as e:
        print(e)
        continue
```

```python
start_time=time.time()
name_list=[]
level_list=[]
address_list=[]
detail_list=[]
star_list=[]
price_list=[]
sold_list=[]
lng_list=[]
lat_list=[]
def get_jwd(address):
    url='http://api.map.baidu.com/geocoding/v3/?'
    params={'output':'json','ak':'2hy78RthBxMMRUWxux2VctP1p71e0Ezj','address':address}
    e_params=urllib.parse.urlencode(params)
    e_url=url+e_params
    r=requests.get(e_url)
    j=json.loads(r.text)
    if j['status']==0:
        lng=j['result']['location']['lng']
        lat=j['result']['location']['lat']
    else:
        lng=''
        lat=''
    return lng,lat

link_list=[]
link1='https://piao.qunar.com/ticket/list.htm?keyword=%E7%83%AD%E9%97%A8%E6%99%AF%E7%82%B9&region=&from=mps_search_suggest&page='
for i in range(50):
    link=link1+str(i+1)
    link_list.append(link)

for i,url in enumerate(link_list):
    print('爬到第{}页'.format(i+1))
    r=requests.get(url,headers=headers)
    time.sleep(random.uniform(1,3))
    soup=BeautifulSoup(r.text,'lxml')
    sight_list=soup.select('#search-list > div')
    for sight in sight_list:
        name=sight.find('h3',class_='sight_item_caption').a['title']
        name_list.append(name)
        try:
            level=sight.find('span',class_='level').text.replace('景区','')
            level_list.append(level)
        except:
            level_list.append('')
        address=sight.find('p',class_='address color999').span['title']
        address_list.append(address)
        lng,lat=get_jwd(address)
        lng_list.append(lng)
        lat_list.append(lat)
        detail=sight.find('div',class_='intro color999').text
        detail_list.append(detail)
        star=sight.find('span',class_='product_star_level').em.span.text.replace('热度','')
        star_list.append(star)
        price=sight.find('span',class_='sight_item_price').em.text
        price_list.append(price)
        sold=sight.find('span',class_='hot_num').text
        sold_list.append(sold)
end_time=time.time()
print('用时：',end_time-start_time)
```

```python
columns=['景点名称','星级','地址','热度','价格','销售量','价绍','经度','纬度']
df=pd.DataFrame(data=[name_list,level_list,address_list,star_list,price_list,sold_list,detail_list,lng_list,lat_list]).T
df.columns=columns
df.to_csv('D:/python3.7/HEHE/输出/热门景点前750个.csv',index=False,encoding='utf_8_sig')
```
