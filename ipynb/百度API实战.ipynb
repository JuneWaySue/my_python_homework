```python
def hehe(loc,page_num=0):
    link='http://api.map.baidu.com/place/v2/search'
    headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.25 Safari/537.36 Core/1.70.3704.400 QQBrowser/10.4.3587.400'}
    pa={'q':'公园',
       'region':loc,
       'scope':'2',
       'page_size':'20',
       'page_num':page_num,
       'coord_type':1,
       'output':'json',
       'ak':'GzAUudGpyI0zqzGRpQ8YmWEsmt0FAj47'}
    r=requests.get(link,headers=headers,params=pa)
    return json.loads(r.text)
```

```python
pro_city=['广东省','山东省','江苏省','福建省','云南省']
for each in pro_city:
    res=hehe(each)
    for temp in res['results']:
        name=temp['name']
        num=temp['num']
        output='\t'.join([name,str(num)])+'\n'
        with open('./citys.txt','a+',encoding='utf-8') as f:
            f.write(output)
            f.close()
```

```python
city_list=[]
with open('./citys.txt','r',encoding='utf-8') as f:
    for i in f:
        c=i.split('\t')
        city=c[0]
        city_list.append(city)
    f.close()
city_list[0]='广州市'
city_list
```

```python
name_list=[]
province_list=[]
address_list=[]
tel_list=[]
lat_list=[]
lng_list=[]
for each in city_list:
    flag=True
    page_num=0
    while flag:
        res=hehe(each,page_num=page_num)
        print("爬到"+each,page_num)
        try:
            if res['results']:
                for eachone in res['results']:
                    name=eachone['name']
                    province=eachone['province']
                    address=eachone['address']
                    try:
                        tel=eachone['telephone']
                    except:
                        tel=None
                    lat=eachone['location']['lat']
                    lng=eachone['location']['lng']
                    name_list.append(name)
                    province_list.append(province)
                    address_list.append(address)
                    tel_list.append(tel)
                    lat_list.append(lat)
                    lng_list.append(lng)
                page_num+=1
            else:
                flag=False
        except:
            flag=False
```

```python
columns=['公园名称','省份','地址','电话','经度','纬度']
df=pd.DataFrame(data=[name_list,province_list,address_list,tel_list,lat_list,lng_list]).T
df.columns=columns
df
```

```python
df.to_excel('./公园.xlsx',index=False)
```
