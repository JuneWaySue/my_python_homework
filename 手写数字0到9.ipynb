## 上机练习：每个数字取10种手写体显示其图像，通过子图方式显示

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
fig=plt.figure(figsize=(10,10),facecolor="gray")
d=datasets.load_digits()
X=d.data
y=d.target
some_digit=[[],[],[],[],[],[],[],[],[],[]]
for i in range(150):
    k=y[i]
    some_digit[k].append(X[i])
i=1
for j in range(10):
    for k in range(10):
        ax=fig.add_subplot(10,10,i)
        some_digit_image=some_digit[j][k].reshape(8,8)
        ax.imshow(some_digit_image,cmap=matplotlib.cm.binary)
        ax.axis("off")
        i=i+1
plt.show()
```

## 特征处理
### 对清洗后的HR表的satisfaction_level，	last_evaluation，	number_project，	average_monthly_hours，	time_spend_company，	Work_accident，	promotion_last_5years	分别进行归一化和标准化处理
### 对department和salary进行标签编码或独热编码

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler,StandardScaler
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
from sklearn.preprocessing import Normalizer
```

```python
df=pd.read_csv("HR.csv",engine='python')
#清洗数据
df=df.dropna(subset=["satisfaction_level","last_evaluation"])
df=df[df["satisfaction_level"]<=1][df["salary"]!="nme"]
```

```python
g1=MinMaxScaler().fit_transform(df["satisfaction_level"].reshape(-1,1))
g2=MinMaxScaler().fit_transform(df["last_evaluation"].reshape(-1,1))
g3=MinMaxScaler().fit_transform(df["number_project"].reshape(-1,1))
g4=MinMaxScaler().fit_transform(df["average_monthly_hours"].reshape(-1,1))
g5=MinMaxScaler().fit_transform(df["time_spend_company"].reshape(-1,1))
g6=MinMaxScaler().fit_transform(df["Work_accident"].reshape(-1,1))
g7=MinMaxScaler().fit_transform(df["promotion_last_5years"].reshape(-1,1))
#归一化得到的新表
d1={"satisfaction_level":list(g1),"last_evaluation":list(g2),"number_project":list(g3),"average_monthly_hours":list(g4),
    "time_spend_company":list(g5),"Work_accident":list(g6),"promotion_last_5years":list(g7)}
df_GYH=pd.DataFrame(d1,index=df.index)
df_GYH.loc[0:10]
```

```python
b1=StandardScaler().fit_transform(df["satisfaction_level"].reshape(-1,1))
b2=StandardScaler().fit_transform(df["last_evaluation"].reshape(-1,1))
b3=StandardScaler().fit_transform(df["number_project"].reshape(-1,1))
b4=StandardScaler().fit_transform(df["average_monthly_hours"].reshape(-1,1))
b5=StandardScaler().fit_transform(df["time_spend_company"].reshape(-1,1))
b6=StandardScaler().fit_transform(df["Work_accident"].reshape(-1,1))
b7=StandardScaler().fit_transform(df["promotion_last_5years"].reshape(-1,1))
#标准化得到的新表
d2={"satisfaction_level":list(b1),"last_evaluation":list(b2),"number_project":list(b3),"average_monthly_hours":list(b4),
    "time_spend_company":list(b5),"Work_accident":list(b6),"promotion_last_5years":list(b7)}
df_BZH=pd.DataFrame(d2,index=df.index)
df_BZH.loc[0:10]
```

```python
q1=LabelEncoder().fit_transform(df["department"].reshape(-1,1))
q2=LabelEncoder().fit_transform(df["salary"].reshape(-1,1))
#标签化编码得到的新表
d3={"department":q1,"salary":q2}
df_BQH=pd.DataFrame(d3,index=df.index)
df_BQH.loc[0:10]
```

```python
#独热编码
lb_encoder=LabelEncoder()
lb_tran_f=lb_encoder.fit_transform(q1)
oht_encoder=OneHotEncoder().fit(lb_tran_f.reshape(-1,1))
oht_tran_f=oht_encoder.transform(lb_encoder.transform(q2).reshape(-1,1))
oht_tran_f.toarray()
```
